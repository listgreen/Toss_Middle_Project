### **1) 수익성 지표 (Profitability Indicators)**
| 컬럼명 | 의미 |
|--------|------|
| **ROA (Return on Assets, 총자산이익률)** | 기업의 총자산 대비 얼마나 많은 이익을 창출했는지 |
| **당기순이익 (Net Income)** | 특정 기간 동안 기업이 벌어들인 최종 이익 |
| **자산총계 (Total Assets)** | 기업이 보유한 전체 자산 |
| **ROE (Return on Equity, 자기자본이익률)** | 자기자본 대비 순이익 비율 |
| **자본총계 (Total Equity)** | 기업의 총 자기자본 |
| **ROIC (Return on Invested Capital, 투자자본수익률)** | 기업이 투자한 자본 대비 창출한 이익 비율 |
| **NOPLAT (Net Operating Profit Less Adjusted Taxes)** | 조정 세후 영업이익 |
| **IC (Invested Capital, 투자자본)** | 기업이 실제로 투자한 자본 |
| **자본금순이익률 (Net Income to Capital Ratio)** | 자본금 대비 순이익 비율 |
| **자본금 (Paid-in Capital)** | 기업이 발행한 주식의 액면가 총액 |
| **매출원가율 (Cost of Sales Ratio)** | 매출액 대비 매출원가 비율 |
| **매출원가 (Cost of Goods Sold)** | 기업이 제품을 생산하는 데 든 비용 |
| **수익 (Revenue)** | 기업의 총 매출액 |

```python
def df_merging(df1, df2):
    df_merged = pd.merge(df1, df2, on='종목코드', how='inner', suffixes=('_df1', '_df2'))
    common_cols = set(df1.columns) & set(df2.columns)
    common_cols.remove("종목코드")
    for col_name in common_cols:
        df_merged[col_name] = df_merged[col_name + "_df1"].combine_first(df_merged[col_name + "_df2"])
        df_merged.drop(columns=[col_name + "_df1", col_name + "_df2"], inplace=True)
    return df_merged
```
